```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load dataset
df = pd.read_csv("C:/Users/xiang/Desktop/Vhack25/CareEase/backend/data/healthcare_chatbot_dataset.csv")

# Convert age to age group
def get_age_group(age):
    if 0 <= age <= 9:
        return "0-9"
    elif 10 <= age <= 19:
        return "10-19"
    elif 20 <= age <= 29:
        return "20-29"
    elif 30 <= age <= 39:
        return "30-39"
    elif 40 <= age <= 49:
        return "40-49"
    elif 50 <= age <= 59:
        return "50-59"
    elif 60 <= age <= 69:
        return "60-69"
    elif 70 <= age <= 79:
        return "70-79"
    elif 80 <= age <= 89:
        return "80-89"
    else:
        return "90+"

# Example user input
user_input_age = 22
age_group = get_age_group(user_input_age)

extracted_info = {
    "Symptoms": ["fever", "loss_of_taste"],
    "Duration": "52 days",
    "Age Group": age_group,
    "Gender": "Male",
    "Medical History": "asthma",
    "Medications": "none",
    "Allergies": ["latex", "penicillin", "seafood"],
    "Biological Traits": "immunocompromised",
    "Lifestyle & History": ["physically_active", "vaccinated"],
}

# Convert user info and dataset rows into flat text
def flatten_info(info):
    flat_text = []
    for key, value in info.items():
        if isinstance(value, list):
            flat_text.extend(value)
        else:
            flat_text.append(str(value))
    return " ".join(flat_text).lower()

user_text = flatten_info(extracted_info)
corpus = df.apply(lambda row: flatten_info(row), axis=1).tolist()
corpus.insert(0, user_text)

# TF-IDF + Cosine Similarity
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)
cosine_similarities = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()

# Find top 3 diseases + combined recommendation
df['similarity_score'] = cosine_similarities
top_matches = df.sort_values(by='similarity_score', ascending=False).drop_duplicates(subset='Disease').head(3)

# Get recommendations from the top 3 diseases
recommendations = top_matches['Recommendations'].tolist()

# Combine all recommendations into one
combined_recommendation = " ".join(recommendations)

# Optionally, clean the combined recommendation to avoid redundancy
# Example: remove repeated common sentences
combined_recommendation_sentences = combined_recommendation.split('. ')
unique_recommendations = list(dict.fromkeys(combined_recommendation_sentences))  # Removing duplicates
final_recommendation = '. '.join(unique_recommendations)

# Print results
print("ðŸ§  Top 3 Disease Predictions:\n")
for idx, row in top_matches.iterrows():
    print(f"ðŸ©º Disease: {row['Disease']} (Score: {row['similarity_score']:.4f})")

print(f"\nâœ… Combined Smart Recommendation:\n{final_recommendation}")
```
